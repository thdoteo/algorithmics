## Exercise 1.1 | Max

Maximum of three numbers.

```python
def max3(x, y, z):
    """Returns maximum of 3 numbers."""
    maxi = x
    if y > maxi:
        maxi = y
    if z > maxi:
        maxi = z
    return maxi
```

## Exercise 1.2 | Next day

Compute date of next day.

```python
def leap(y):
    """Tests whether the year is leap."""
    return (y % 4 == 0) and ((y % 100 != 0) or (y % 400 == 0))

def nb_days(m, y):
    """Number of days in a month."""
    if m == 2:
        return 28 + leap(y)
    else:
        return 30 + (m % 7) % 2 + (m == 7)
    
def valid_date(d, m, y):
    """Tests whether date is valid."""
    return (0 < m and m < 13) and (0 < d <= nb_days(m, y))

def next_day(d, m, y):
    """Computes next day's date."""
    nbdays = nb_days(m, y)
    d = (d % nbdays) + 1
    if d == 1:
        m = (m % 12) + 1
        if m == 1:
            y = y + 1
    return d, m, y 

# main 
d = int(input("Give me a date, days to start : "))
m = int(input("Next is month : "))
y = int(input("What year?"))

if not valid_date(d, m, y):
    raise Exception("Date is note valid.")
else:   
    print("Date of next day is : ", next_day(d, m, y))
```

## Exercise 2.1 | Compute list to 9

```python
def mirror(n):
    """Computes mirror of a 2-digit number."""
    return (n % 10)*10 + (n // 10)

def abs(n):
    """Computes absolute value of entry."""
    if n > 0:
        return n
    else:
        return -n
    
def list_to_nine(n):
    """Prints list to nine of a 2-digit number."""
    if n < 10:
        print(n)
    else:
        print(n, end=", ")
        n = abs(n - mirror(n))
        list_to_nine(n)

# main
n = int(input("Give a 2-digit positive number : "))
if n > 99 or n < 10:
    raise Exception("Not a 2-digit number.")
else:
    list_to_nine(n)
    
```

```python
def _sumdivisors(n, p, sqrt):
    """Computes sum of divisors of n bigger or equal to p."""
    if p > sqrt:
        print()
        return 0
    if p == sqrt:
        print(p, end=" ")
        return p
    else:
        if n % p == 0:
            print(p, end=" ")
            res = p + (n // p) + _sumdivisors(n, p + 1, sqrt)
            print(n // p, end=" ")
            return res
        else:
            return _sumdivisors(n, p + 1, sqrt)

def sumdivisors(n):
    """Computes sum of proper divisors of n."""
    sqrt = math.sqrt(n)
    print(1, end=" ")
    return 1 + _sumdivisors(n, 2, sqrt)

n = int(input("Give me a positive integer : "))
if n < 1:
    raise Exception("This is not a positive integer.")
else:
    print("The list of divisors of entry is : ", end="")
    res = sumdivisors(n)
    print("\nThe fact", n, "is a perfect number is : ", n == res)     
    
```
